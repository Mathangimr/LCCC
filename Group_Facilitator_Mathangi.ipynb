```python
Cols_having_nan=loan_data.isna().sum() ##Nan - not an integer, indicating missing values, 
#There are around columns having all records as Nans, those columns do not contribute much for the analysis, hence we can drop them

print(type(Cols_having_nan))
print(Cols_having_nan)
```

```python
##Getting list of columns where all records are null

print(loan_data.columns[loan_data.isnull().all(0)].size) ##54 columns are totally nulls and there is no use in our analysis
print(loan_data.columns[loan_data.isnull().all(0)])
```

```python
loan_data.agg(['count', 'size', 'nunique'])
```

```python
loan_data1=loan_data.dropna(axis=1,how='all')
loan_data1.shape
```

```python
loan_data1.isnull().all(axis=1).count()
```

```python
# function to get the percentage of null values
def get_null_value(loan_data1):
    op=round(loan_data1.isnull().sum()/len(loan_data1)*100, 2)
    return op
    
```

```python
Cols_nulls=get_null_value(loan_data1) 
Cols_nulls
##There are few columns where there are 100% null values and columns where more than 50% of the values are null, hence removing them as well
```

```python
##Removing columns where there is >50% of nulls as they are not useful in data analysis.
Cols_nulls_50=Cols_nulls[Cols_nulls>50]
print(Cols_nulls_50)
print(len(Cols_nulls_50.index))
```

```python
loan_data1.drop(Cols_nulls_50.index, axis=1, inplace=True)
loan_data1.shape
```

```python
# id is considered as index instead of member_id, because if the same member is applying for a new loan, then member_id might be same but id will be a new id given for that member
## Setting up index is useful for faster runtime, slicing and manipulation
# id column is unique for all rows
loan_data1.set_index('id', inplace=True)
```
